```python
import zipfile
import os

def extrair_arquivos_especificos_de_pasta_zip(pasta_zipada, caminhos_arquivos, pasta_de_destino):
    # Verifica se a pasta zipada existe
    if not os.path.isfile(pasta_zipada):
        print(f'A pasta zipada não existe: {pasta_zipada}')
        return

    with zipfile.ZipFile(pasta_zipada, 'r') as zip_ref:
        for caminho_arquivo in caminhos_arquivos:
            try:
                zip_ref.extract(caminho_arquivo, pasta_de_destino)
                print(f'Arquivo extraído: {caminho_arquivo}')
            except KeyError:
                print(f'Arquivo não encontrado na pasta zipada: {caminho_arquivo}')

# Exemplo de uso da função
pasta_zipada = 'ETLSIM.zip'  # Pasta ZIP que contém os arquivos
caminhos_arquivos = caminhos_arquivos  # Lista de caminhos dos arquivos a extrair
pasta_destino = 'SIM_2018_ate_2020'  # Pasta para onde os arquivos serão copiados

extrair_arquivos_especificos_de_pasta_zip(pasta_zipada, caminhos_arquivos, pasta_destino)
```

```python
import os

def obter_caminhos_arquivos(pasta):
    caminhos_arquivos = []
    
    # Verifica se a pasta existe
    if os.path.exists(pasta):
        # Obtém a lista de arquivos na pasta
        arquivos_na_pasta = os.listdir(pasta)
        
        # Obtém caminho completo para cada arquivo
        caminhos_arquivos = [os.path.join(pasta, arquivo) for arquivo in arquivos_na_pasta]
    else:
        print(f"A pasta {pasta} não existe.")

    return caminhos_arquivos

# Exemplo de uso
pasta_alvo = "SIM_2018_ate_2020"
caminhos_dos_arquivos = obter_caminhos_arquivos(pasta_alvo)

# Exibir caminhos dos arquivos
for caminho in caminhos_dos_arquivos:
    print(caminho)
```

```python
def retira_colunas_e_linhas_nan_de_dataframes_e_salva(lista_caminhos, pasta_destino):

    for caminho in lista_caminhos:

        # Carrega o DataFrame a partir do arquivo CSV
        df = pd.read_csv(caminho,sep=',')

        # Remove colunas com todos os valores NaN
        df = df.dropna(axis=1, how='all')

        # Remove linhas com valores NaN
        df = df.dropna(axis=0, how='all')

        # Remove linhas duplicadas
        df = df.drop_duplicates()

        # Extrai o nome do arquivo do caminho
        nome_arquivo = os.path.basename(caminho)

        # Caminho de destino para salvar o DataFrame processado
        caminho_destino = os.path.join(pasta_destino, nome_arquivo)

        # Salva o DataFrame processado como arquivo CSV
        df.to_csv(caminho_destino, index=False)


pasta_destino = 'SIM_2018_ate_2020_processados'
retira_colunas_e_linhas_nan_de_dataframes_e_salva(caminhos_dos_arquivos, pasta_destino)
```

```python
def colunas_em_comum_entre_dataframes(caminhos):
    # Lista para armazenar os DataFrames
    dataframes = []

    # Carrega os DataFrames a partir dos caminhos fornecidos
    for caminho in caminhos:
        df = pd.read_csv(caminho)
        dataframes.append(df)

    # Obter as colunas em comum
    if dataframes:
        colunas_em_comum = set(dataframes[0].columns).intersection(*(df.columns for df in dataframes))

        if colunas_em_comum:
            return list(colunas_em_comum)
        else:
            return None
    else:
        return None

# Exemplo de uso
pasta_alvo = "SIM_2018_ate_2020_processados"
caminhos_dos_arquivos = obter_caminhos_arquivos(pasta_alvo)

# Chamar a função e obter a lista de colunas em comum
lista_colunas_em_comum = colunas_em_comum_entre_dataframes(caminhos_dos_arquivos)

# Exibir a lista de colunas em comum
if lista_colunas_em_comum:
    print("Colunas em comum entre os DataFrames:")
    print(lista_colunas_em_comum)
else:
    print("Não há colunas em comum entre os DataFrames ou nenhum DataFrame foi fornecido.")
```

# 3. Concatenação de DataFrames

```python
def obter_caminhos_arquivos(pasta):
    caminhos_arquivos = []
    
    # Verifica se a pasta existe
    if os.path.exists(pasta):
        # Obtém a lista de arquivos na pasta
        arquivos_na_pasta = os.listdir(pasta)
        
        # Obtém caminho completo para cada arquivo
        caminhos_arquivos = [os.path.join(pasta, arquivo) for arquivo in arquivos_na_pasta]
    else:
        print(f"A pasta {pasta} não existe.")

    return caminhos_arquivos

# Exemplo de uso
pasta_alvo = "SIM_2018_ate_2020_processados"
caminhos_dos_arquivos = obter_caminhos_arquivos(pasta_alvo)

# Exibir caminhos dos arquivos
for caminho in caminhos_dos_arquivos:
    print(caminho)
```

```python
import os
import pandas as pd

def concatena_dfs(caminhos_dos_arquivos, caminho_pasta_destino):
    # Lista para armazenar os DataFrames
    lista_dataframes = []

    # Carrega os DataFrames a partir dos caminhos fornecidos
    for caminho in caminhos_dos_arquivos:
        df = pd.read_csv(caminho)  # Você pode ajustar o método de leitura dependendo do formato dos seus DataFrames
        lista_dataframes.append(df)

    # Concatena os DataFrames ao longo das linhas
    df_merged = pd.concat(lista_dataframes, ignore_index=True)

    # Salva o DataFrame concatenado em um arquivo CSV
    caminho_completo = os.path.join(caminho_pasta_destino, "SIM_concatenada.csv")
    df_merged.to_csv(caminho_completo, index=False)

# Exemplo de chamada da função com caminhos dos DataFrames

caminho_pasta_destino = 'SIM_concatenada'

concatena_dfs(caminhos_dos_arquivos, caminho_pasta_destino)

```
