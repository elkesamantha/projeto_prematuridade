```python
import zipfile
import os
import pandas as pd
```

```python
df_SIM = pd.read_csv('SIM_concatenada\SIM_concatenada.csv', sep = ',')
```

```python
df_SINASC = pd.read_csv('SINASC_FINAL.csv', sep = ',')
```

```python

def colunas_em_comum_sem_nan(df1, df2):
    # Obter conjuntos de colunas que não têm NaN em ambos os DataFrames
    colunas_sem_nan_df1 = set(coluna for coluna in df1.columns if not df1[coluna].isna().any())
    colunas_sem_nan_df2 = set(coluna for coluna in df2.columns if not df2[coluna].isna().any())

    # Encontrar as colunas em comum sem NaN
    colunas_em_comum_sem_nan = colunas_sem_nan_df1 & colunas_sem_nan_df2

    return colunas_em_comum_sem_nan

colunas_comum_sem_nan = colunas_em_comum_sem_nan(df_SIM, df_SINASC)
colunas_comum_sem_nan = list(colunas_comum_sem_nan)

print("Colunas em comum sem NaN:")
print(colunas_comum_sem_nan)
```

```python


df_SIM_chaves = df_SIM[colunas_comum_sem_nan]
df_SINASC_chaves = df_SINASC[colunas_comum_sem_nan]
```
